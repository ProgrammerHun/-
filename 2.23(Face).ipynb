```python
import matplotlib.patches as patches # 패치 

img = mping.imread('face1.jpg')

fig, ax = plt.subplots(figsize = (10,10))  # 프레임
ax.imshow(img)

rect_face = patches.Rectangle((x,y),w,h, linewidth = 5, edgecolor = 'r', facecolor  = 'none')
ax.add_patch(rect_face)

plt.text(10,40, annotation, wrap = True, fontsize=17, color='white')
plt.show()
```

```python
import requests  # 인터넷 연결

client_id = "4jY20csjiavpBsFwUQhG"
client_secret = "6dMMUa62jl"

url = "https://openapi.naver.com/v1/vision/face"
files = {'image':open('family.jpg','rb')}
headers = {'X-Naver-Client-Id':client_id, 'X-Naver-Client-Secret': client_secret}

response = requests.post(url, files = files, headers = headers)

import json # json 형식을 다루는 모듈
parsed = json.loads(response.text)
print(json.dumps(parsed, indent = 4, sort_keys = False, ensure_ascii=False))
```

```python
import matplotlib.patches as patches # 패치 

img = mping.imread('family.jpg')

fig, ax = plt.subplots(figsize = (10,10))  # 프레임
ax.imshow(img)

for each in parsed['faces']:
    x,y,w,h = each['roi'].values()
    gender,gen_confi = each['gender'].values()
    emotion, emo_confi = each['emotion'].values()
    age, age_confi = each['age'].values()

    rect_face = patches.Rectangle((x,y),w,h, linewidth = 5, edgecolor = 'r', facecolor  = 'none')
    ax.add_patch(rect_face)
    
    annotation = gender + " : " + str("%0.2f%%" % gen_confi) + '\n' +  emotion + ' : ' + str("%0.2f%%" % emo_confi) + '\n' + age + ' : ' + str("%0.2f%%" % age_confi)


    
    plt.text(x-10,y-200, annotation, wrap = True, fontsize=10, color='black')
   

    

plt.show()


```
